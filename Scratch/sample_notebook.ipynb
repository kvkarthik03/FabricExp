````python
import matplotlib.pyplot as plt
from skimage import io, transform, color, draw
import pandas as pd
````

Read an image: ```io.imread```

Display an image: ```io.imshow()```

Resolution of image: ```img.shape```

````python
image = io.imread('/Users/nitinrajesh/Desktop/1441826.jpg')

#image = io.imread('/Users/nitinrajesh/Code/FabricExp/op_3.jpeg')

io.imshow(image)

print(image.shape)

small_img = transform.rescale(image, 0.25)
transform.resize()
print(len(small_img.flatten()))

df = pd.DataFrame(small_img.flatten())

df.to_excel('Pixel_vals.xlsx',index=False)

# image = transform.rescale(image, 0.1)
# image = transform.resize(102,85)

# image_arr = (image.flatten())
# print(image_arr)
# len(image_arr)

````

````python
img_hsv = color.rgb2hsv(image)
img_rgb = color.hsv2rgb(img_hsv)
img_gry = color.rgb2gray(image)
img_recolor = color.gray2rgb(img_gry)
img_lab = color.rgb2lab(image)
img_yuv = color.rgb2yuv(image)

io.imshow(img_hsv)
io.imshow(img_rgb)
io.imshow(img_yuv)
print(img_recolor.shape)
````

````python
img_yiq = color.rgb2yiq(image)
io.imshow(img_yiq)
````

````python
x,y = draw.line(0,0,1000,1000)
l,b,c = image.shape
u,v = draw.line(2000,2000,l-1,b-1)
image[x,y] = (200,0,200)
image[u,v] = (50,200,50)
io.imshow(image)
````

````python
row, col = draw.polygon([1000,2000,1000,2000],[2000,2000,1000,1000])
image[row,col] = (250,200,000)
io.imshow(image)
````

````python
x,y = draw.ellipse(200,200,c_radius=100,r_radius=100)

image[x,y] = 1

io.imshow(image)
````
